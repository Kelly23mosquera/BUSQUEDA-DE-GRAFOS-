```python
# Programa que implementa los algoritmos BFS y DFS
# Para recorrer un grafo y comparar el recorrido
# Y la memoria utilizada por cada algoritmo.

from collections import deque # deque es una estructura de datos tipo cola que permite agregar y quitar elementos rápidamente.
import sys  # para medir el tamaño en memoria de las estructuras



# Definición del grafo usando un diccionario.
# Cada nodo del grafo tiene una lista de nodos vecinos a los que está conectado.
grafo = {
    'P': ['Q', 'R'],
    'Q': ['S', 'T'],
    'R': ['U'],
    'S': [],
    'T': ['U'],
    'U': []
}

# ALGORITMO BFS (Busqueda en amplitud)

# Función que implementa la búsqueda en anchura
def bfs(grafo, inicio):

    # Conjunto para almacenar los nodos que ya fueron visitados
    visitados = set()

    # Cola que se utilizará para recorrer el grafo por niveles
    cola = deque([inicio])

    # Lista donde se guardará el orden del recorrido
    recorrido = []

    # Mientras haya nodos en la cola
    while cola:

        # Extrae el primer nodo de la cola
        nodo = cola.popleft()

        # Verifica si el nodo no ha sido visitado
        if nodo not in visitados:

            # Marca el nodo como visitado
            visitados.add(nodo)

            # Guarda el nodo en la lista de recorrido
            recorrido.append(nodo)

            # Recorre todos los vecinos del nodo actual
            for vecino in grafo[nodo]:

                # Si el vecino aún no fue visitado se agrega a la cola
                if vecino not in visitados:
                    cola.append(vecino)

    # Calcula una estimación de la memoria usada por las estructuras del algoritmo
    memoria_usada = sys.getsizeof(cola) + sys.getsizeof(visitados) + sys.getsizeof(recorrido)

    # Devuelve el recorrido realizado y la memoria usada
    return recorrido, memoria_usada



# ALGORITMO DFS (Busqueda en profundida)

# Función que implementa la búsqueda en profundidad
def dfs(grafo, inicio, visitados=None, recorrido=None):

    # Si visitados es None se crea un nuevo conjunto
    if visitados is None:
        visitados = set()

    # Si recorrido es None se crea una nueva lista
    if recorrido is None:
        recorrido = []

    # Marca el nodo inicial como visitado
    visitados.add(inicio)

    # Guarda el nodo en el recorrido
    recorrido.append(inicio)

    # Recorre todos los vecinos del nodo actual
    for vecino in grafo[inicio]:

        # Si el vecino no ha sido visitado se llama recursivamente a DFS
        if vecino not in visitados:
            dfs(grafo, vecino, visitados, recorrido)

    # Calcula la memoria usada por el algoritmo
    memoria_usada = sys.getsizeof(visitados) + sys.getsizeof(recorrido)

    # Devuelve el recorrido y la memoria utilizada
    return recorrido, memoria_usada


# EJECUCIÓN DEL PROGRAMA

# Ejecuta BFS empezando desde el nodo p
recorrido_bfs, memoria_bfs = bfs(grafo, 'P')

# Ejecuta DFS empezando desde el nodo p
recorrido_dfs, memoria_dfs = dfs(grafo, 'P')

# Muestra en pantalla el recorrido realizado por BFS
print("Recorrido BFS:", recorrido_bfs)

# Muestra la memoria usada por BFS
print("Memoria usada en BFS:", memoria_bfs, "bytes")

# Imprime una línea en blanco para separar resultados
print("\nRecorrido DFS:", recorrido_dfs)

# Muestra el recorrido realizado por DFS
print("Memoria usada en DFS:", memoria_dfs, "bytes")
```
