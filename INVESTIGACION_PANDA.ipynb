## ¿Qué es Pandas?

Pandas es una librería de Python diseñada para el análisis y manipulación de datos. Su principal función es facilitar el trabajo con información estructurada mediante herramientas como los DataFrames, que permiten organizar datos en filas y columnas de manera eficiente.

## ¿Para qué sirve?

Pandas sirve para leer, organizar, limpiar, filtrar y analizar datos provenientes de diferentes fuentes, como archivos CSV, Excel y bases de datos. Además, permite realizar operaciones estadísticas y transformar información de forma rápida, facilitando el análisis de grandes volúmenes de datos.

## Objetivo principal de la librería

El objetivo principal de Pandas es proporcionar herramientas eficientes para el procesamiento y análisis de datos estructurados en Python. Esta librería busca simplificar tareas complejas relacionadas con la manipulación de información, permitiendo trabajar con datos de manera más organizada y precisa.

## Principales funciones y características

Entre las principales funciones y características de Pandas se encuentran la creación y manejo de DataFrames, lectura y escritura de archivos CSV y Excel, filtrado y ordenamiento de datos, operaciones entre columnas, manejo de datos faltantes y realización de análisis estadísticos. Además, se caracteriza por ser una herramienta rápida, flexible y fácil de integrar con otras librerías de Python.

Pandas es fundamental en la inteligencia artificial y el análisis de datos porque permite preparar y organizar la información antes de utilizar modelos predictivos o algoritmos de aprendizaje automático. ##  Creación de un DataFrame utilizando la librería Pandas

En este ejercicio se crea un DataFrame utilizando información de una librería ficticia, incluyendo títulos de libros, autores, precios y cantidad disponible en stock.

```python
import pandas as pd  # Importar la librería Pandas

# Datos de una librería ficticia
titulos = ["Cien años de soledad", "El Principito", "Don Quijote de la Mancha", "Rayuela"]  # Lista de títulos de libros

autores = ["Gabriel García Márquez", "Antoine de Saint-Exupéry", "Miguel de Cervantes", "Julio Cortázar"]  # Lista de autores

precios = [45000, 30000, 38000, 42000]  # Lista de precios en pesos colombianos

stock = [12, 20, 8, 5]  # Cantidad de libros disponibles

# Crear el DataFrame con la información de los libros
df_libros = pd.DataFrame({
    "Título": titulos,
    "Autor": autores,
    "Precio (COP)": precios,
    "Stock disponible": stock
})

# Mostrar tabla organizada
df_libros
```

## Ejercicio 2: Operaciones entre columnas

En este ejercicio se realizan operaciones matemáticas entre columnas para calcular el promedio de notas y la diferencia entre exámenes de varios estudiantes.

```python
import pandas as pd  # Importar la librería Pandas

# Crear un DataFrame con los nombres de estudiantes y sus notas
df_estudiantes = pd.DataFrame({
    "Nombre": ["Ana", "Luis", "Carlos", "María"],  # Nombres de los estudiantes
    "Examen 1": [3.5, 4.2, 2.8, 4.8],  # Notas del primer examen
    "Examen 2": [4.0, 3.9, 3.5, 4.6]   # Notas del segundo examen
})

# Calcular el promedio de las dos notas
df_estudiantes["Promedio"] = (
    df_estudiantes["Examen 1"] + df_estudiantes["Examen 2"]
) / 2

# Calcular la diferencia entre el segundo y el primer examen
df_estudiantes["Diferencia"] = (
    df_estudiantes["Examen 2"] - df_estudiantes["Examen 1"]
)

# Mostrar el DataFrame con los nuevos resultados
df_estudiantes
```

## Ejercicio 3: Lectura y análisis de un archivo CSV

Este código utiliza la librería Pandas para leer un archivo CSV desde una URL publicada en Google Sheets. Luego organiza la información eliminando columnas vacías y asignando nombres correctos a las columnas para mostrar los datos climáticos de forma estructurada.

```python
import pandas as pd  # Importar la librería Pandas

# URL del archivo CSV publicado desde Google Sheets
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT5_GnSWDCCplPpAe0slIV3BqlZaWLW4_dUgCljU1Ye51u2uBQv878nbe3QNbaM-Q/pub?gid=552832994&single=true&output=csv"

# Leer el archivo CSV y omitir las primeras filas innecesarias
df_clima = pd.read_csv(url, skiprows=6)

# Eliminar columnas completamente vacías
df_clima = df_clima.dropna(axis=1, how='all')

# Asignar nombres correctos a las columnas
df_clima.columns = ["Fecha", "Temperatura", "Humedad", "Lluvia"]

# Mostrar las primeras filas del DataFrame
df_clima.head()
```

## Ejercicio 4: Trabajo con datos sintéticos

En este ejercicio se generarán datos aleatorios utilizando NumPy para simular información de ventas de una tienda.

```python
import pandas as pd  # Importar la librería Pandas
import numpy as np   # Importar NumPy para generar datos aleatorios

# Generar datos sintéticos de ventas entre 50 y 500
ventas = np.random.randint(50, 500, 10)

# Generar datos sintéticos de clientes entre 5 y 50
clientes = np.random.randint(5, 50, 10)

# Crear un DataFrame con la información generada
df_ventas = pd.DataFrame({
    "Ventas diarias": ventas,
    "Clientes atendidos": clientes
})

# Mostrar el DataFrame en forma de tabla
df_ventas
```
